```python
import requests
from bs4 import BeautifulSoup

# 设置要爬取的PTT板块和关键词
board = 'KoreaStar'
keyword = 'IVE'

# 设置要爬取的页数
pages = 10
```

```python
import pandas as pd
#容器
data = pd.DataFrame(columns=['title', 'article_url'])
#水瓢將資料一筆一筆貼上去
temp = pd.DataFrame(columns=['title', 'article_url'])
```

```python
# 爬取每一页的文章列表
for page in range(1, pages + 1):
    url = f'https://www.ptt.cc/bbs/{board}/search?q={keyword}&page={page}'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
    
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"无法访问页面 {url}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')
    articles = soup.find_all(class_='r-ent')

    # 输出每页的文章标题和链接
    for article in articles:
        title = article.find(class_='title')
        if title is not None:
            title_text = title.text.strip()
            article_url = 'https://www.ptt.cc' + title.a['href']
            print(title_text)
            print(article_url)
            temp = {'title': title_text, 'article_url': article_url}
            data = data.append(temp, ignore_index = True)
```
